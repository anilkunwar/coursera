### Autocovariance function

#### Random variables

$$ X : S \rightarrow \mathbb{R} $$

where $S$ is the sample space of the experiment.

#### Covariance

$X, Y$ are two random variables. Then, **covariance** measures the linear dependence between two random variables

$$ Cov(X, Y) = \mathbb{E}[(X-\mu_X)(Y - \mu_Y)] = Cov(Y, X) $$

#### Stochastic Processes

Collection of a random variables, $X_1, X_2, X_3, \dots $

$$ X_t \sim distribution(\mu_t, \sigma_t^2 )$$

#### Autocovariance function

Covariance of different elements in our stochastic process.

$$ \gamma(s, t) = Cov(X_s, X_t) = \mathbb{E}[(X_s - \mu_s)(X_t - \mu_t)] $$
$$ \gamma(t, t) = \mathbb{E}[(X_t - \mu_t)^2] = Var(X_t) = \sigma_t^2 $$

Usually, we denote its autocovariance function like this,

$$\gamma_k = \gamma(t, t + k) \approx c_k$$

Actually, this notation can be used in the condition of stationary time series. 

### Autocovariance coefficients

#### Estimation of the covariance

Once we have a paired dataset,

$$ (x_1, y_1), (x_2, y_2), \dots, (x_N, y_N) $$

The estimation of covariance is,

$$s_{xy} = \frac{\sum_{t=1}^N (x_t - \bar{x})(y_t - \bar{y})}{N - 1} $$

#### Autocovariance coefficients

Autocovariance coefficients at different lags

$$ \gamma_k = Cov(X_t, X_{t+k}) $$

$c_k$ is an estimation of $\gamma_k$.

We assume (weak) stationary

#### Estimation

$$ c_k = \frac{\sum_{t=1}^{N-k}(x_t - \bar{x})(x_{t+k} - \bar{x})}{N} $$

where

$$ \bar{x} = \frac{\sum_{t=1}^N x_t}{N} $$

```r
purely_random_process = ts(rnorm(100))
```

```r
print(purely_random_process)
```

```r
(acf(purely_random_process, type='covariance'))
```

### Autocorrelation Function (ACF)

- We assume weak stationarity

- The autocorrelation coefficient between $X_t$ and $X_{t+k}$ is defined to be

$$ -1 \lt \rho_k = \frac{\gamma_k}{\gamma_0} \lt 1 $$

- Estimation of autocorrelation coefficient at lag $k$

$$ \begin{aligned} r_k &= \frac{c_k}{c_0} \\
  &= \frac{\sum_{t=1}^{N-k}(x_t - \bar{x})(x_{t+k} - \bar{x})}{\sum_{t=1}^N(x_t - \bar{x})^2} \end{aligned}$$

```r
(acf(purely_random_process, main='Correlogram of a purely random process'))
```
