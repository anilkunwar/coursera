### Simple Linear Regression

#### The Mauna Loa CO2 Data

```r
help(co2)
```

```r
plot(co2, main='Atmospheric CO2 Concentration')
```

The response (i.e. co2 concentration) of the ith observation may be denoted by the random variable $Y_i$ This response depends upon the explanatory variable $x_i$ in a linear way, with some noise added as,

$$ Y_i = (\beta_0 + \beta_1 x_i) + \epsilon_i $$

Error term $\epsilon_i$ is:
 - measurement error
 - lack of knowledge of other important influences,
 - etc...

(often reasonable!) assumptions
- the errors are normally distributed and on average zero
- the errors all have the same variance (they are homoscedastic),
- the errors are unrelated to each other (they are independent across observations)

#### Ordinary Least Square (OLS)

$$ Q = \sum(\text{observed} - \text{predicted})^2 \\
 Y_i = i^{th} \text{ observed response variable} \\
 \begin{aligned} \hat{Y_i} &= i^{th} \text{ predicted response variable} \\
 &= \text{ slope} \cdot x_i + \text{intercept} \end{aligned} $$

#### Linear model of co2

```r
(co2.linear.model = lm(co2 ~ time(co2)))
```

```r
summary(co2.linear.model)$coefficients
```

```r
plot(co2, main='Atmospheric CO2 Concentration with Fitted Line')
abline(co2.linear.model, col='red')
```

### Inference

#### The Gosset Data

```r
help(sleep)
```

```r
plot(extra~group, data=sleep, main="Extra Sleep in Gosset Data by Group")
attach(sleep)
extra.1 = extra[group == 1]
extra.2 = extra[group == 2]
```

#### T-distribution Test

```r
t.test(extra.1, extra.2, paired=TRUE, alternative="two.sided")
```

These data are highly significant ( $\text{p-value} \lt 0.05$)

#### Unpack this data

$H_0$: Mean response is the same for both drugs, or

$$ H_0: \mu_{drug1} - \mu_{drug2} \equiv \mu_{diff} = 0 $$

$H_1$: Mean response is not the same for both drugs, or

$$H_1: \mu_{drug1} - \mu_{drug2} \equiv \mu_{diff} \neq 0 $$

#### Unpack this output

$$ \alpha \equiv P(\text{Type I Error}) = 0.05 $$

$$ t = \frac{\bar{d} - 0}{S_d \over \sqrt{n}} = \frac{-1.58 - 0}{1.229995483 \over \sqrt{10}} = -4.062127683$$

$\bar{d} \equiv \text{ average of differences} = \text{ difference of averages} \\
 S_d \equiv \text{ standard deviation of differences} \\
 n \equiv \text{ sample size} $

#### Get tail area of twice in two-sided t test for p-value

```r
p = 2 * pt(-4.062127683, 9)
p
```

$$ p \lt \alpha \Rightarrow \text{ reject} \\
 p \gt \alpha \Rightarrow \text{ do not reject} $$

#### General Framework for Hypothesis Tests
- State clearly what your variables are (define your terms)
- State the null and alternative hypothesis
- Decide upon a level of significance, $\alpha$
- Compute a test statistic ($z, t, \chi^2$ and $F$ are popular stats)
- find the p-value corresponding to your test statistic (for left/right/or two tailed test)
- Form a conclusion: if $p \lt \alpha$ (improbable data) reject $H_0$, otherwise do not reject. We typically do not accept, just like the courts never say that someone is innocent.

#### Confidence interval

A common form for a confidence interval

$$ Estimate \pm Table value \cdot (Estimated) Standard Error \\
 \bar{d} \pm t_{\alpha \over 2} \cdot \frac{s}{\sqrt{n}}$$
