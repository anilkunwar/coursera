## Histogram in R

### Objective
- To obtain the histogram of a dataset
- To impose smooth density function over a histogram
- To change colors and line width in a histogram

```r
small.size.dataset=c(91,49,76,112,97,42,70, 100, 8, 112, 95, 90, 78, 62, 56, 94, 65, 58, 109, 70, 109, 91, 71, 76, 68, 62, 134, 57, 83, 66)
```

```r
hist(small.size.dataset)
```

```r
hist(small.size.dataset, xlab='My data points')
```

```r
hist(small.size.dataset, xlab='My data points', main='Histogram of my data')
```

#### From frequency to Density

```r
hist(small.size.dataset, xlab='My data points', main='Histogram of my data', freq=FALSE)
```

#### Change the color

```r
hist(small.size.dataset, xlab='My data points', main='Histogram of my data', freq=FALSE, col='green')
```

```r
hist(small.size.dataset, xlab='My data points', main='Histogram of my data', freq=FALSE, col='green')
lines(density(small.size.dataset))
```

```r
hist(small.size.dataset, xlab='My data points', main='Histogram of my data', freq=FALSE, col='green')
lines(density(small.size.dataset), col='red', lwd=5)
```

```r
hist(small.size.dataset, xlab='My data points', main='Histogram of my data', freq=FALSE, col='green', breaks=10)
lines(density(small.size.dataset), col='red', lwd=5)
```

## Scatterplot in R

### Objective
- To obtain scatterplot for a bivariate data

```r
set.seed(2016)
```

```r
Test_1_scores=round(rnorm(50, 78, 10))
Test_2_scores=round(rnorm(50, 70, 14))
```

```r
Test_1_scores
```

```r
Test_2_scores
```

```r
plot(Test_2_scores~Test_1_scores)
```

```r
plot(Test_2_scores~Test_1_scores, main='Test scores for two exams (50 students)', xlab='Test 1 scores', ylab='Test 2 scores')
```

```r
plot(Test_2_scores~Test_1_scores, main='Test scores for two exams (50 students)', xlab='Test 1 scores', ylab='Test 2 scores', col='blue')
```
